#### 1. Написать программу, оценивающую значение машинного эпсилон для стандартных чисел с двойной точностью.

```python
k = 0
x2 = np.float64(1)
y2 = np.float64(2)
while y2 != np.float64(1):
    x2 /= np.float64(2)
    y2 = np.float64(1) + x2
    k += 1
print('Машинное эпсилон с двойной точностью' ,2**(-k+1))
```

#### 3. Релизовать алгоритм вычисления первой производной методом дифференцирования с комплексным шагом. Для функции с известным значением производной, изучить сходимость численного ответа при уменьшении шага.

Вычислим численно значение производной $cos'(x)=-sin(x)$ в точке $\pi/4$

```python
x = np.linspace(-0.75, 1.5, 100)
y = np.cos(x)
plt.plot(x, y)
plt.plot(np.pi / 4, np.cos(np.pi / 4), 'o')
```

```python
# Function of numerical differentiation with complex step
def deriv_complex(f, x, h):
    return( f(x + 1j*h).imag / h )
```

```python
# Comparison of derivative values with different steps
for i in range(1, 11):
    h = 10**(-i)
    print('derivative=',deriv_complex(np.cos, np.pi/4, h), 'h=', h)
print('           ', -np.sin(np.pi/4), '<--- true value of derivative')
```

#### 5. Написать программу, вычисляющую производную функции $x^2 \ln x$ в точке $x = 0$ с точностью $O(h^2)$.
